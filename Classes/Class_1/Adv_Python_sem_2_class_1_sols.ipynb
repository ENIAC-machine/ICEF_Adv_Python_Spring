### Concept of deep copy and shallow copy

In programming sometimes you need to store the same information in several variables. Because memory is scarce and variables are many programmers have come up with a good workaround for using the same data several times. Instead of copying the data into a variable, you copy a `pointer` (or a link in non-nerd language) to the original data. This is called `shallow copy` because you don't copy the data itself, you just copy the **reference** to the data. Proper copying of the data is called `deep copy`. Although shallow copy can be very useful, it is also dangerous as if you change the original data the data in the variables holding `shallow copies` will also change. See the picture and the example below

![image](deep_vs_shallow_copy_python.jpg)

````python
list_a = [1, 2, 15] #initialize a list
print(f'Original list_a: {list_a}')

list_b = list_a #put the list into another variable, not list_b points to list_a
print(f'list_b: {list_b}')

list_c = list_a #do the same for another variable
print(f'list_c: {list_c}')

print('\nAfter we change list_a...\n')

#Now we change something in list_a
list_a[2] = 3

#And we see that ALL the lists that point to list_a change with it, although we didn't change them specifically
print(f'list_a: {list_a}')
print(f'list_b: {list_b}')
print(f'list_c: {list_c}')
````

## Loops

In programming you will need to apply a single (at times complex) operation to a group of values *very* often. For that there are `loops`- tools to help apply the same set of operations/ instructions to several elements (often organized in some sort of compound structure like a list). In python there are 2 kinds of loops: the `for` loop and the `while` loop

### The `for` loop 

To use the for loop in python you need to apply the `for` statement to some `Iterable` - that is an object through which you can iterate like a `list`, a `tuple` or a `str` (you iterate through characters)

Suppose you have a list `nums` with integer values

````python
nums = [1, 5, 6, 19]
````

And now suppose you want to print each value with some fancy text around it. Instead of doing it the inefficient way by *manually referencing each value and printing it*. You can just use the `for` loop 

````python
#below the variable element takes turns storing values from nums. At first it stores value 1, then 5 and so on until the `nums` list is exhausted
for element in nums:
    print(f'Look at this value -> {element} !!!')
````

Now you may assume that all the usage of the `for` loop is this trivial

Let's examine a common incorrect use case for the for loop:

    You have a list of integers, which you need to increment by some value, how would you do that with a for loop?

Some of you would write something like this:

````python
nums = [1, 5, 6, 19]
value = 5

for element in nums:
    element = element + value #incrementing
````

Let's see if anything has changed in the `nums` list

````python
print(nums)
````

But why? Everything seems right, we took a number then incremented it. Well here is the trick, you didn't take the value that was in the `nums` array, you took **it's copy**. That's right, the `for` loops operate with copies of the values from the `Iterable`, not with the originals. How do we correctly increment the values then?

First, you need to measure the length of the list via the `len` function, which takes `Iterable` as an argument and returns the number of elements it contains:

````python
len_nums = len(nums)
print(len_nums)
````

Now use the python's `range` function in the form `range(N)` to list values from 0 to N-1, which will correspond to the indices of the list

````python
for index in range(len_nums):
    nums[index] += value #equivalent to nums[index] = nums[index] + value

print(nums)
````

### The `while` loop

The `while` loop goes while the user-specified condition holds, for example, I have an array of numbers:

```python
nums = [5, 6, 17, 10]
```

And I want to print each number until I encounter a number above 15, below you'll see a way to do that with a while loop

````python
nums = [5, 6, 17, 19]

#introduce the variable to index the values in the list
index = 0

#set the condition
while nums[index] <= 15:
    print(nums[index]) #print the number
    index += 1 #equivalent to index = index + 1
````

You must be careful as sometimes the while loop can be infinite:

```python

a = 0
while a < 1:
    a -= 1 #equivalent to a = a - 1
```
The loop above will run forever as the variable `a` will *always* be smaller than 1

## Functions

As your code gets bigger you will find it difficult to name the variables and also you will find that you are repeating yourself often. Functions were designed specifically to combat these 2 issues, they also represent the most fundamental principle in programming called `DRY` for `Don't Repeat Yourself`. We will try to uphold this principle in this course too. Obviously, you shouldn't be fanatical about it, but try your best not to repeat yourself

What exactly is a function? It is a half-isolated `space` in the code, where you can write code separately (by default) from the rest of the script. All the variables local to the function we call the `local variables`. All the variables, that are defined outside of any function are called `global`. Let's define our first function

````python
def example(): #function named example
    print('Hi there, I am a function!')
````

But why didn't anything happen? Functions should be called for the code inside of them to run. Let's try calling this function

````python
example()
````

Obviously the functions without any communication with the main program are next to useless that's you can pass some parameters into the function. Suppose we need to create a function to multiply 2 numbers and then get the result beck to the main program, let's see the function for that

````python
def multiplication(num1, num2): #num1 and num2 are expected to be provided upon calling the function
    res = num1 * num2
    return res #return the value to the main program
````

Because the function expects num1 and num2 from the user, calling the function without any parameters will result in error

````python
multiplication()
````

Let's try calling the function as intended

````python
num1, num2 = 5, 10 

multiplication(num1, num2)

print('No error!')
````

But where is our result value? To get it we need to assign the output of the function to a variable

````python
num1, num2 = 5, 10

res = multiplication(num1, num2)

print(res)
````

When arguments are many and the code is plently it is hard to understand, what types of values should be in each argument, that's why there are `type hints`, the sole purpose of which is to address this issue, here is the syntax of them using the example of our last function 

````python
def multiplication(num1: int, #num1 is expected to be of type integer
                   num2: int #num2 is also expected to be an integer
                  ) -> int: #the function returns an integer value
    res = num1 * num2
    return res #return the value to the main program
````
