## Functions reminder

A function is a part of the program that we write when we don't want to repeat ourselves with the same snippet of code. Its contents aren't accessible to the namespaces above it, but are accessible to the namespaces below it

### Ayo, wait, what is a namespace?  

So you have a script with some names in it like 

```python
num1 = 15
num2 = 16
```

The names of the variables (and other objects) make up the `global` namespace. With functions you can create `local` namespaces that are visible *exclusively* to the function and the objects inside it. See examples below

````python
#define variables in the global namespace

num1 = 15
num2 = -10
````

````python
#create a function with its own namespace
def example():
    float1 = 15.13
    float2 = -11.01
    
#call the function to execute
example()
````

````python
#try to print the variables defined in the function namespace
print(float1, float2) #this will give an error
````

````python
#define a new function that referes to the global namespace
def example2():
    print(num1, num2)

#call it
example2()
````

So you see that the function has access to the namespace *above* it and doesn't have access to the namespace *below* it. It works within functions too!

````python
#define a function with its own namespace
def example3():

    num3 = 18

    #now inside it define 2 more functions with their own namespaces contained inside the namespace of example3

    def subexample1():
        num4 = 100
        print(num1, num2, num3)

    def subexample2():
        num5 = -100
        print(num1, num2, num3)

    subexample1()
    subexample2()

#call the function
example3()
````

As your code gets more complex and more functions emerge it becomes hard to remember what types of data are expected as inputs. Luckily, python has a way for you to indicate explicitly what datatypes you want to input in the function:

```python
def example_func(num1: int,
                 num2: int
                ) -> float:
    
    return float(num1 + num2)
```

In the function above we say that we expect argument num1 to be of `integer` type as well as num2 and that the function return a variable of type `float`. These type hints are **not** checked by the python interpreter, but serve as a piece of advice for your fellow programmers and yourself

Functions also allow you to pass the default values for parameters, see example below

````python
#define a function with two parameters, out of which one has a default value
def example_default(num1: int, num2: int = 5) -> int:
    return num1 + num2

#now when we call it, we don't have to pass 2 parameters, we can just pass the first one and the second one will be filled by a default value
example_default(1)
````

There are also special functions called `generators`. They are designed when you want the function to *generate* values as a sequence. For the function to become a generator you need to replace the `return` expression with the `yield` expresion. You iterate over the generator via the `next()` function. When the generator runs out of values, it raises an error. See the example below

````python
#define a generator
def example_gen():
    yield 15

a = example_gen()
print(next(a)) #this will output 15
print(next(a)) #this will return an error
````

Example of use in a for loop

````python
#define a generator
def example_gen():
    #yield numbers from 0 to 9 (inclusive)
    for i in range(10):
        yield i

gen = example_gen()

for num in gen:
    print(num)
    if num > 5:
        break

print('----------')

for other in gen:
    print(other)
````

You can also transform generators into lists:

````python
#define a generator
def example_gen_2(num: int):
    for i in range(num):
        yield i

list_nums = list(example_gen_2(10))
list_nums
````

Write a function `find_max_min` with one argument `numbers`- list that returns both maximum and minimum values from a list without using max() or min() (returns them as a tuple)

````python
def find_max_min(numbers: list) -> tuple[float, float]:
    max_val, min_val = numbers[0], numbers[0]
    for num in numbers:
        if num > max_val:
            max_val = num
        if num < min_val:
            min_val = num

    return max_val, min_val

numbers = [100, 200, 5, 16, 1241, 68912]
find_max_min(numbers)
````

### Exercise 29

Write a function that takes a string and returns a dictionary counting the frequency of each character (without using collections.Counter). Include type hints

````python
def freq_counter(text: str) -> dict:
    chars = list(set(text))
    counter = dict()
    for char in chars:
        counter.update({char : 0})

    for char in text:
        counter[char] += 1

    return counter

freq_counter('Hello there!')
````

### Exercise 32

Create a generator function `fibonacci_generator` that accepts one argument (`n` - integer) and yields the first `n` Fibonacci numbers one at a time. Use yield to generate values lazily. Don't use the type hints to indicate the output type.

````python
def fibonacci_generator(n: int):
    
    def fib(n):
        if n == 1:
            return 0
        elif n == 2:
            return 1
        else:
            return fib(n-1) + fib(n-2)

    for i in range(1, n+1):
        yield fib(i)


for i in fibonacci_generator(4):
    print(i)
````
