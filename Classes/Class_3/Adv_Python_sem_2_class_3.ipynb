## What is pandas?

The `pandas` library is the primary tool in Python for manipulating, cleaning, and analyzing structured data. It essentially adds "spreadsheet" or "SQL-like" capabilities to Python, allowing you to work with tabular data efficiently and intuitively.

Here is a short (approximate) list of reasons for why it is used:

*   **Handling Tabular Data**: I mean this one is self-explanatory
*   **Performance**: Built on top of **NumPy**, it is much faster than standard Python lists for large datasets because it uses optimized C-based operations under the hood. 
*   **Integration**: It works seamlessly with other key libraries like **Matplotlib** (for plotting), **Scikit-learn** (for machine learning), and **SQL** databases, acting as the central hub for data pipelines.
*   **Versatile I/O**: It can read and write data from many file formats with a single line of code, including CSV, Excel, JSON, and SQL databases.

## Pandas primary datatypes

`pandas` has 2 primary datatypes: `pd.DataFrames` and `pd.Series`. They are used to work with tables and series of data respectively, we can (and often do) construct them from the python-native datatypes, below are all the ways to do that

```python
# --- 1. Creating Series ---

# From a List
s_list = pd.Series([10, 20, 30, 40], name='FromList')

# From a Dictionary
s_dict = pd.Series({'a': 100, 'b': 200, 'c': 300}, name='FromDict')

# From a Scalar value (requires index)
s_scalar = pd.Series(5, index=['a', 'b', 'c', 'd'], name='FromScalar')


# --- 2. Creating DataFrames ---

# From a Dictionary of Lists
dict_of_lists = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago']
}
df_dict_lists = pd.DataFrame(dict_of_lists)

# From a List of Dictionaries
list_of_dicts = [
    {'Name': 'David', 'Age': 40, 'City': 'Miami'},
    {'Name': 'Eve', 'Age': 45, 'City': 'Seattle'}
]
df_list_dicts = pd.DataFrame(list_of_dicts)

# From a List of Lists (requires column names)
list_of_lists = [
    ['Frank', 'Engineering'],
    ['Grace', 'Marketing']
]
df_list_lists = pd.DataFrame(list_of_lists, columns=['Employee', 'Department'])

# From a List of Tuples
list_of_tuples = [
    ('Item1', 10.5),
    ('Item2', 20.0)
]
df_tuples = pd.DataFrame(list_of_tuples, columns=['Product', 'Price'])

# From a Dictionary of Series
dict_of_series = {
    'col1': pd.Series([1, 2, 3], index=['a', 'b', 'c']),
    'col2': pd.Series([4, 5, 6, 7], index=['a', 'b', 'c', 'd'])
}
df_dict_series = pd.DataFrame(dict_of_series)


# Displaying results
print("--- Series ---")
print("From List:\n", s_list, "\n")
print("From Dict:\n", s_dict, "\n")
print("From Scalar:\n", s_scalar, "\n")

print("--- DataFrames ---")
print("From Dict of Lists:\n", df_dict_lists, "\n")
print("From List of Dicts:\n", df_list_dicts, "\n")
print("From List of Lists:\n", df_list_lists, "\n")
print("From List of Tuples:\n", df_tuples, "\n")
print("From Dict of Series:\n", df_dict_series, "\n")
```

## File interaction with pandas

`pandas` allows you to save the `DataFrames` you work with into files, we will be interested in working with the `.csv` and `.xlsx` formats in particular. Below you can find examples of saving a `DataFrame` into a file

```python
# Create a sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago']
}
df = pd.DataFrame(data)

# --- 1. Saving to CSV ---

# Basic save
df.to_csv('users.csv', index=False)
# index=False prevents pandas from writing row numbers (0, 1, 2) to the file

# Custom separator (e.g., pipe instead of comma)
df.to_csv('users_pipe.csv', sep='|', index=False)

# Saving without headers
df.to_csv('users_no_header.csv', header=False, index=False)


# --- 2. Saving to Excel (.xlsx) ---
# Note: You may need to install the 'openpyxl' library first, put this in your command line: pip install openpyxl

# Basic save
df.to_excel('users.xlsx', index=False)

# Save with specific sheet name
df.to_excel('users_sheet.xlsx', sheet_name='Employees', index=False)

# Saving to multiple sheets in one file
with pd.ExcelWriter('multiple_sheets.xlsx') as writer:
    df.to_excel(writer, sheet_name='Sheet1', index=False)
    df.to_excel(writer, sheet_name='Sheet2', index=False)
```

## Inspecting DataFrames

There will be many times, when you'll need to access a specific value in the `DataFrame`, there are 2 main ways to do this: 

```python
# DataFrame with custom index (labels) for clarity
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [24, 27, 22, 32, 29],
    'City': ['NY', 'LA', 'Chicago', 'Houston', 'Phoenix'],
    'Salary': [70000, 80000, 65000, 120000, 95000]
}

df = pd.DataFrame(data)
df.set_index('Name', inplace=True)

print("--- Original DataFrame ---")
print(df)
print("\n" + "-"*40 + "\n")


# ==========================================
# .loc[] -> Label-based (uses Index names and Column names)
# Syntax: df.loc[row_label, column_label]
# ==========================================

print("--- .loc[] (Label-based) ---")

print(f"1. Bob's Age: {df.loc['Bob', 'Age']}")

print(f"\n2. Charlie's data:\n{df.loc['Charlie']}")

print(f"\n3. Multiple rows (Alice, David):\n{df.loc[['Alice', 'David']]}")

print(f"\n4. Row Slice (Bob to David):\n{df.loc['Bob':'David']}")

print(f"\n5. Sub-table (Bob:David, City:Salary):\n{df.loc['Bob':'David', ['City', 'Salary']]}")

print(f"\n6. People older than 25:\n{df.loc[df['Age'] > 25]}")


print("\n" + "-"*40 + "\n")


# ==========================================
# .iloc[] -> Integer-based (uses numeric positions 0, 1, 2...)
# Syntax: df.iloc[row_position, column_position]
# ==========================================

print("--- .iloc[] (Position-based) ---")

# 1. Single cell: Row 1 (Bob), Column 0 (Age) -> Remember index 'Name' is not column 0 anymore
# Columns are: Age(0), City(1), Salary(2)
print(f"1. Element at row 1, col 0 (Bob's Age): {df.iloc[1, 0]}")

# 2. Single row: Get row 2 (Charlie)
print(f"\n2. Row 2 (Charlie):\n{df.iloc[2]}")

# 3. Multiple rows: Get row 0 and row 3 (Alice and David)
print(f"\n3. Rows 0 and 3:\n{df.iloc[[0, 3]]}")

# 4. Slicing rows: From 1 to 3 (EXCLUSIVE of end index in .iloc!)
# This gets rows 1 (Bob) and 2 (Charlie), but NOT 3 (David)
print(f"\n4. Row Slice 1:3 (Bob, Charlie):\n{df.iloc[1:3]}")

# 5. Sub-table: Rows 1:3, Columns 0:2 (Age and City)
print(f"\n5. Slice Rows 1:3 and Cols 0:2:\n{df.iloc[1:3, 0:2]}")

```

Also to understand what data you're dealing with you'll need to extract some aggregate values from the `DataFrame` like the mean of some of its columns etc. to sum up the functions at your disposal:

`.mean()` - Calculate mean

`.median()` - Calculate median

`.sum()` - Calculate sum

`.max()` - Find maximum

`.min()` - Find minimum

`.std()` - Calculate standard deviation

`.groupby()` - Group data

`.describe()` - Generate descriptive statistics

Below you can see the example usages of the the functions above funtions

```python
#A sample DataFrame
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'C', 'C', 'B'],
    'Sales': [100, 200, 150, 250, 120, 300, 320, 210],
    'Profit': [10, 25, 15, 30, 12, 40, 45, 20]
}
df = pd.DataFrame(data)

print("--- Original DataFrame ---")
print(df)
print("-" * 30)

# 2. Basic Statistical Functions (Aggregate whole columns)
# Note: numeric_only=True is often required if non-numeric columns exist
print(f"Mean Sales: {df['Sales'].mean()}")
print(f"Median Profit: {df['Profit'].median()}")
print(f"Total Sales (Sum): {df['Sales'].sum()}")
print(f"Max Profit: {df['Profit'].max()}")
print(f"Min Sales: {df['Sales'].min()}")
print(f"Std Dev of Sales: {df['Sales'].std():.2f}")
print("-" * 30)

# 3. Descriptive Statistics
print("--- Descriptive Statistics (.describe()) ---")
# Generates count, mean, std, min, 25%, 50%, 75%, max for numeric cols
print(df.describe())
print("-" * 30)

# 4. Grouping Data (.groupby())
# Group by 'Category' and calculate the mean for each group
print("--- Group by Category (Mean) ---")
grouped_mean = df.groupby('Category').mean()
print(grouped_mean)

print("\n--- Group by Category (Sum) ---")
grouped_sum = df.groupby('Category').sum()
print(grouped_sum)

print("\n--- Group by Category (Multiple Aggregations) ---")
# You can also apply multiple functions at once using .agg()
grouped_agg = df.groupby('Category')['Sales'].agg(['mean', 'sum', 'max'])
print(grouped_agg)
```

## Combining DataFrames

If you want to combine 2 (or more) `DataFrames`, then you will most likely `join` them. To `join` two `DataFrames` is to select the the rows of both via a certain principle, in total there are `4` basic principles: 

1) The `inner` join - select only the values, which are present in *both* `DataFrames`
2) The `outer` join - select **all** values (from both `DataFrames`)
3) The `left` and `right` joins - select the values that are in the left/ right `DataFrame`

Below you can see the graph and a meme to help you remember how joining works

![image](join_meme.jpg)

![image](SQL_JOIN_VIS.png)

To perform the `join` operation, pandas has the `merge` operation (and a method for a `DataFrame`)


Below are the examples of uses of the methods to join 2 `DataFrames`

```python
#'DepartmentID' is the common column
df_employees = pd.DataFrame({
    'EmployeeID': [1, 2, 3, 4, 5],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'DepartmentID': [101, 102, 101, 103, 104]
})

df_departments = pd.DataFrame({
    'DepartmentID': [101, 102, 103, 105],
    'DeptName': ['HR', 'IT', 'Marketing', 'Finance']
})

print("--- Employees ---")
print(df_employees)
print("\n--- Departments ---")
print(df_departments)
print("-" * 30)


# ==========================================
# pd.merge()
# ==========================================

# A. INNER JOIN
# Keeps only rows where the key exists in BOTH DataFrames.
# Eve (Dept 104) is lost because 104 isn't in Departments.
# Finance (Dept 105) is lost because no employee is there.
df_inner = pd.merge(df_employees, df_departments, on='DepartmentID', how='inner')
print("\n--- INNER JOIN (Only matching records) ---")
print(df_inner)

# B. LEFT JOIN
# Keeps ALL rows from the Left DF (Employees). 
# If a match isn't found in Right DF, it fills with NaN (Eve will have NaN DeptName).
df_left = pd.merge(df_employees, df_departments, on='DepartmentID', how='left')
print("\n--- LEFT JOIN (All Employees) ---")
print(df_left)

# C. RIGHT JOIN
# Keeps ALL rows from the Right DF (Departments).
# If a match isn't found in Left DF, it fills with NaN (Finance will have NaN Employee info).
df_right = pd.merge(df_employees, df_departments, on='DepartmentID', how='right')
print("\n--- RIGHT JOIN (All Departments) ---")
print(df_right)

# D. OUTER JOIN (Full Outer)
# Keeps ALL rows from BOTH sides. Matches where possible, NaNs where not.
df_outer = pd.merge(df_employees, df_departments, on='DepartmentID', how='outer')
print("\n--- OUTER JOIN (Everything) ---")
print(df_outer)
```
